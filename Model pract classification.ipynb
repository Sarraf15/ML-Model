```python
cat_col=['BFPcase','Gender','BMIcase']
```

```python
#changing the dataype from object to int
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for i in cat_col:
    df[i]=le.fit_transform(df[i])
```

```python
df.head(2)
```

```python
#splitting the data into dependent and independent varaibles
x=df.drop('Exercise Recommendation Plan',axis=1)
y=df['Exercise Recommendation Plan']
```

```python
#training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score
```

```python
def eval_mod(ytest,ypred):
    print('classification_report \n',classification_report(ytest,ypred))
    print('confusion_matrix \n',confusion_matrix(ytest,ypred))
    print('accuracy_score \n',accuracy_score(ytest,ypred))

def mscore(model):
    print('Training score:',model.score(x_train,y_train))
    print('Testing score:', model.score(x_test,y_test))
```

k-Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()

knn.fit(x_train,y_train)
```

```python
mscore(knn)
```

Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
mscore(dtc)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
mscore(rfc)
```

```python
ypred_rfc=rfc.predict(x_test)
ypred_rfc[:10]
```

```python
eval_mod(y_test,ypred_rfc)
```
