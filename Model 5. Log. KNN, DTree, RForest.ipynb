collecting all 'object' datatype column in one place

```python
obj_coll=['workclass','education','marital.status','occupation','relationship','race','sex','native.country','income']
```

label encoding to change the datatypes from 'object' to 'int' or change from 'text' column to 'int' colum for model building

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for i in obj_coll:
    df[i]=le.fit_transform(df[i])
```

```python
df.head(2)
```

drop 'fnlwgt' column , which is of no use

```python
#dropping 'fnlwgt' column 
df.drop('fnlwgt',axis=1,inplace=True)
```

```python
df.head(2)
```

splitting the data into dependant and independant variables

```python

x=df.drop('income',axis=1)
y=df.income
```

```python
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix ,accuracy_score

def eval(ytest,ypred):
    print('classification_report \n',classification_report)
    print('confusion_matrix \n',confusion_matrix)
    print('accuracy_score \n',accuracy_score)

def mscore(model):
    print('Training Score: ',model.score(x_train,y_train))
    print('Testing score: ',model.score(x_test,y_test))
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
mscore(lr)
```

KNN CLASSIFICATION

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
mscore(knn)
```

DECISION TREE CLASSIFIER

```python
from sklearn.tree import  DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',max_depth=3,min_samples_split=7)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=80,criterion='gini',max_depth=5,min_samples_split=11)
rf.fit(x_train,y_train)
```

```python
mscore(rf)
```
