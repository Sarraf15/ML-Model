```python
df['latitude'] =df.fillna(df['latitude'].mean(),inplace=True)
df['longitude'] =df.fillna(df['longitude'].mean(),inplace=True)
df['trim'] = df.fillna(df['trim'].mode(),inplace=True)
df['area_name'] = df.fillna(df['area_name'].mode(),inplace=True)
df['engine_capacity_cc '] = df.fillna(df['engine_capacity_cc'].mode(),inplace=True)
df['horsepower'] = df.fillna(df['horsepower'].mode(),inplace=True)
```

```python
cat_col=['brand', 'model', 'trim', 'kilometers', 'year',
       'vehicle_age_years', 'regional_specs', 'doors', 'body_type',
       'fuel_type', 'seating_capacity', 'transmission_type',
       'horsepower', 'no_of_cylinders', 'exterior_color',
       'interior_color', 'warranty',  'country', 'city', 'area_name',
        'latitude', 'longitude', 'seller_type',
       'engine_capacity_cc ']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for col in cat_col:
    df[col]=le.fit_transform(df[col])
```

```python
df1=df.copy()
```

```python
#droping adress
df1.drop(['address','location_name','engine_capacity_cc'],axis=1,inplace=True)
df1.head()
```

```python
corr=df1.corr()
sns.heatmap(corr,annot=True)
```

```python
#splitting into dependant andd independent variables
x=df1.drop('price',axis=1)
y=df1.price
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score
```

```python
def eval(ytest,ypred):
    cm=confusion_matrix(ytest,ypred)
    print(cm)
    print('accuracy_score',accuracy_score(ytest,ypred))
    print('classification_report',classification_report(ytest,ypred))

def mscore(model):
    print('training score:',model.score(x_train,y_train))
    print('testing score:',model.score(x_test,y_test))
```

```python
#importing linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

lr.fit(x_train,y_train)
```

```python
mscore(lr)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)

knn.fit(x_train,y_train)
```

```python
mscore(knn)
```

```python
from sklearn.tree import DecisionTreeClassifier
tr=DecisionTreeClassifier(criterion='gini')

tr.fit(x_train,y_train)
```

```python
mscore(tr)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()

dtr.fit(x_train,y_train)
```

```python
mscore(dtr)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10)

rf.fit(x_train,y_train)
```

```python
mscore(rf)
```
