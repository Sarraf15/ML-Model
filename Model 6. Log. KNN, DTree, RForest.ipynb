```python
import pandas as pd
df=pd.read_csv(r"C:\Users\shubh\Downloads\datasets_Fish - datasets_Fish.csv.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
#changing the datatype of target variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Species']=le.fit_transform(df['Species'])
```

```python
df.head()
```

spilitting the data into dependent and independent varible

```python
x=df.drop('Species',axis=1)
y=df['Species']
```

Scaling the independent data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

training and testing 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=.2)
print(len(x_train))
print(len(x_test))
```

bulding the model

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

creating a fuction to check the accuracy of model

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
def eval_model(ytest,ypred):
    print('classification_report\n',classification_report(ytest, ypred))
    print('confusion_matrix\n',confusion_matrix(ytest,ypred))
    print('accuracy_score\n',accuracy_score(ytest,ypred))
    

def mscore(model):
    print('Training score', model.score(x_train,y_train))
    print('Tresting score', model.score(x_test,y_test))
```

```python
mscore(knn)
```

```python
#predicting the values
ypred_knn=knn.predict(x_test)
```

```python
eval_model(y_test,ypred_knn)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
mscore(lr)
```

```python
ypred_lr=lr.predict(x_test)
```

```python
eval_model(y_test,ypred_lr)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt=dt.predict(x_test)
```

```python
eval_model(y_test,ypred_dt)
```

Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rd=RandomForestClassifier()
rd.fit(x_train,y_train)
```

```python
mscore(rd)
```

```python
ypred_rd=rd.predict(x_test)
```

```python
eval_model(y_test,ypred_rd)
```

comparing the model based on their accuracy

```python
pred={'Logistic':accuracy_score(y_test,ypred_lr),
      'KNN':accuracy_score(y_test,ypred_knn),
      'DTree':accuracy_score(y_test,ypred_dt),
      'RForest':accuracy_score(y_test,ypred_rd)}
result=pd.DataFrame(pred,index=['Accuracy'])
result.T
```

From the above table , it can be said that RForest is the best but RForest is overfitted model. So the best model is Logistic.
