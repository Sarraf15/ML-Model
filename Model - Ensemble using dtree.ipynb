```python
import pandas as pd

df=pd.read_csv(r"C:\Users\shubh\Downloads\Glass.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Type'].value_counts()
```

```python
#splitting the data into dependent and independent variable
x=df.drop('Type',axis=1)
y=df['Type']
```

```python
#train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
def eval(ytest,ypred):
    print('classification_report\n',classification_report(ytest,ypred))
    print('confusion_matrix \n',confusion_matrix(ytest,ypred))
    print('accuracy_score \n',accuracy_score(ytest,ypred))

def mscore(model):
    print('Training score:',model.score(x_train,y_train))
    print('Testing score:',model.score(x_test,y_test))
```

```python
#importing ensamble
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

dt=DecisionTreeClassifier(criterion='gini',max_depth=3,min_samples_split=9)

ada_boost=AdaBoostClassifier(estimator=dt, n_estimators=40)
ada_boost.fit(x_train,y_train)
```

```python
mscore(ada_boost)
```

```python
y_pred=ada_boost.predict(x_test)
print(y_pred)
```

```python
#evaluating the model
eval(y_test,y_pred)
```

```python
#save the model
import pickle
```

```python
pickle.dump(ada_boost,open('ada_boost.pkl', 'wb'))          #write binary
```

```python
#load the model
loaded_model=pickle.load(open('ada_boost.pkl', 'rb'))       #read binary
```

```python
test_acc=loaded_model.score(x_test,y_test)
print(test_acc)
```
