```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df1=df.copy()
```

```python
cat_col=['Geography','Gender']
```

```python
#label encoding to change the 'object' datatype to 'int' datatype
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for i in cat_col:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

```python
#splitting the data into dependent and independent variables
x=df.drop('Exited',axis=1)
y=df.Exited
```

```python
#now training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.metrics import classification_report, confusion_matrix , accuracy_score

def eval(ytest,ypred):
    print('classification_report \n',classification_report(ytest,ypred))
    print('confusion_matrix \n',confusion_matrix(ytest,ypred))
    print('accuracy_score \n',accuracy_score(ytest,ypred))

def mscore(model):
    print('training score:',model.score(x_train,y_train))
    print('testing score:',model.score(x_test,y_test))
```

```python
#model buliding
#logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

lr.fit(x_train,y_train)
```

```python
mscore(lr)
```

```python
#knn model

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=9)

knn.fit(x_train,y_train)
```

```python
mscore(knn)
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='gini',max_depth=3,min_samples_split=4)

dtc.fit(x_train,y_train)
```

```python
mscore(dtc)
```

```python
y_pred_dtc=dtc.predict(x_test)
y_pred_dtc
```

```python
eval(y_test,y_pred_dtc)
```

```python
#random forst 
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=9)

rfc.fit(x_train,y_train)
```

```python
mscore(rfc)
```
