```python
cat_columns=['region_code',
'segment',                             
'model',                               
'fuel_type',                           
'is_esc',                              
'is_adjustable_steering',              
'is_tpms',                             
'is_parking_sensors',                 
'is_parking_camera',                  
'rear_brakes_type',                   
'transmission_type',                   
'steering_type' ,                    
'is_front_fog_lights',                 
'is_rear_window_wiper',
'is_rear_window_washer',            
'is_rear_window_defogger',           
'is_brake_assist',                     
'is_power_door_locks',                 
'is_central_locking',               
'is_power_steering',                  
'is_driver_seat_height_adjustable',    
'is_day_night_rear_view_mirror',       
'is_ecw',                              
'is_speed_alert']
```

```python
#changing the datatype of 'object'
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()



for i in cat_columns:
    df[i]=le.fit_transform(df[i])
    
```

```python
#showing 2 row of dataframe
df.head(2)
```

```python
#splitting the data into dependent and independent variable
x=df.drop('claim_status',axis=1)
y=df['claim_status']
```

```python
#creating function to evaluate model
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
def eval_model(ytest,ypred):
    print('classification_report\n',classification_report(ytest,ypred))
    print('confusion_matrix\n',confusion_matrix(ytest,ypred))
    print('accuracy_score\n',accuracy_score(ytest,ypred))

def mscore(model):
    print('Training score:',model.score(x_train,y_train))
    print('Testing score:',model.score(x_test,y_test))
```

```python
#training and testing the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1)
print(len(x_train))
print(len(x_test))
```

Building the model

KNeighbors

```python
#kn neighbor
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
mscore(knn)
```

```python
#predicting values
y_pred_knn=knn.predict(x_test)
```

```python
eval_model(y_test,y_pred_knn)
```

Decision Tree 

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
#predicting values
y_pred_dt=dt.predict(x_test)
```

```python
eval_model(y_test,y_pred_dt)
```

Random Forest

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=80,criterion='gini')
rf.fit(x_train,y_train)
```

```python
mscore(rf)
```

```python
#predicting values
y_pred_rf=rf.predict(x_test)
```

```python
eval_model(y_test,y_pred_rf)
```

To find which model is best fitted, we can make a dataframe to get the detail at one place

```python
Pred={'KNN':accuracy_score(y_test,y_pred_knn),
      'DTree':accuracy_score(y_test,y_pred_dt),
      'RForest':accuracy_score(y_test,y_pred_rf)}
result=pd.DataFrame(Pred,index=['Accuarcy'])
result.T
```

From above data we can conclude that KNN is the best model
